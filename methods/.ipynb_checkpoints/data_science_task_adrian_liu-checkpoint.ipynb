#### Definitions
*Following definitions have been taken from the resource links above.*
- Prevalence: A measure of the frequency of a disease of health condition
- Achievement: The QOF gives an indication of the overall achievement of a practice through a points system. Reference to 100 per cent achievement often refers to the percentage of available QOF points achieved.
- Exceptions: Exception reporting’ refers to the potential removal of these patients from calculations of practice achievement for specific clinical indicators.
- Personalised Care Adjustment (PCA): As of 1 April 2019, exception reporting was replaced with Personalised Care Adjustments (PCAs). Unlike exception reporting, PCAs are extracted in the same order for all indicators, enabling more robust insight to be gained from their analysis.
- CVDPP001: In those patients with a new diagnosis of hypertension aged 30 or over and who have not attained the age of 75, recorded between the preceding 1 April to 31 March (excluding those with pre-existing CHD, diabetes, stroke and/or TIA), who have a recorded CVD risk assessment score (using an assessment tool agreed with the NHS CB) of ≥20% in the preceding 12 months: the percentage who are currently treated with statins.

### Reading data to Pandas DataFrame

A pandas dataframe will be our preferred data structure, and to convert from our existing Excel Worksheet structure I require some wrangling/formatting.
1. Parsing of header
2. Selecting slice of worksheet which contains corresponding data

```python
wb = load_workbook('data/qof-1920-prev-ach-pca-cv-prac.xlsx')
ws = wb['CVD-PP']
```

```python
'''
Parsing of the header.
Header spans three rows of the worksheet
Header contains merged rows on row 12
'''
ws['A12'] = None # remove 'Return to Contents' link from header row
header_cols = []
init_col = ''
for cells in zip(ws[12], ws[13], ws[14]):
    if cells[0].value:
        init_col = col = str(cells[0].value)
    elif type(cells[0]) is openpyxl.cell.cell.MergedCell:
        col = init_col
    else:
        col = ''
    for cell in [cells[1], cells[2]]:
        val = cell.value
        if val:
            col += str(val)
    header_cols.append(col)
```

```python
'''
Obtaining worksheet slice for dataframe data structure conversion.
'''
ws.delete_rows(6735, ws.max_row) # delete empty rows at end of file
ws.delete_rows(0, 14) # delete header rows
data = ws.values
df = pd.DataFrame(ws.values, columns=header_cols) # dataframe conversion
```

```python
'''
Add complete column name for "Year on year change".
'''
cols = list(df.columns)
cols[13] = 'Prevelance Year on year change (percentage point)'
cols[18] = 'Achievement Year on year change (percentage point)'
cols[25] = 'PCA Achievement Year on year change (percentage point)'
df.columns = cols
```

```python
def summarise_col(col):
    '''
    Returns a summary stats for a pandas series
    '''
    nunique = col.nunique()
    if col.dtype != 'O':
        mean = col.mean()
        std = col.std()
        min_ = col.min()
        max_ = col.max()
        return nunique, mean, std, min_, max_
    else:
        return nunique, np.nan, np.nan, np.nan, np.nan
```

```python
'''
Display relavent properties and stats for each column
'''
df_summary = df.apply(summarise_col, axis=0).transpose()
df_summary.columns = ['nunique', 'mean', 'std', 'min', 'max']
pd.concat([df_properties, df_summary], axis=1)
```

### Exploratory Data Analysis

First we look at some correlations, mainly we can look at Prevalence, Achievement and Exclusions/PCA across 2018-19, 2019-20 and CVDPP001.

```python
def corr_plot(df):
    '''
    Creates a heatmap for a correlation matrix for a visual representation
    of correlation matrix.
    '''
    corr = df.corr()
    mask = np.triu(np.ones_like(corr, dtype=bool))
    cmap = sns.diverging_palette(200, 0, as_cmap=True)
    plot = sns.heatmap(corr,
                      mask=mask,
                      cmap=cmap,
                      vmax=1,
                      center=0,
                      square=True,
                      linewidths=1)
    annot_plot = sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.5)],
                      mask=mask,
                      cmap=cmap,
                      vmax=1,
                      center=0,
                      square=True,
                      linewidths=1,
                      annot=True,
                      cbar=False)
    return corr, plot, annot_plot
```

```python
_, plot, annot_plot = corr_plot(df.iloc[:, 3:])
plot;
annot_plot;
```

We can notice:
- High Positive correlation with CVDPP001 Achievement and Underlying Achievement net of PCAs with 2019-20 Achievement. Similarly with Denominators.
    - NOTE: CVDPP001PCA Rate and 2019-2020Overall PCA Rate, CVDPP001PCAs and 2019-20Total PCAs, CVDPP001Denominator and 2019-20Total Denominations, CVDPP001Achievement Score and 2019-20Achievement variable pairings have all correlation of 1 and thus we can omit redundent columns.
- Negative correlation with CVDPP001 Patients receiving Intervention and PCA Rates and Total PCAs, but not as much with Exceptions.
- General Positive correlations with many variables which may indicate multicollinearity we might be vary of if used in predictive models.
- We may expect that Overall Exception Rate and Total Exceptions would have a higher correlation.

### Clustering

```python
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import scale
```

```python
'''
Extracting numeric data and scaling.
'''
X = df.iloc[:, 3:]
scaled_X = X.copy()
scaled_X[:] = scale(X)
```

```python
'''
Impute missing values with median.
'''
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
scaled_X[:] = imputer.fit_transform(scaled_X)
```

```python
'''
Fit kmeans model.
'''
kmeans = KMeans(n_clusters=8)
kmeans.fit(scaled_X)
```

```python
'''
Bar plots of counts of clusters splitting by PCN code counts.
'''
col = 'PCN ODS code'
_counts = df[col].value_counts()
_cumsum = _counts.cumsum()
third, two_third = len(_cumsum[_cumsum < int(_cumsum[-1] * (1/3))]),  len(_cumsum[_cumsum < int(_cumsum[-1] * (2/3))])

plt.rcParams['figure.figsize'] = [20, 6]
fig, ax =plt.subplots(1,3)
predicted = kmeans.predict(scaled_X[df[col].isin(_counts.index[0:third])])
sns.countplot(x=predicted, ax=ax[0]);
predicted = kmeans.predict(scaled_X[df[col].isin(_counts.index[third + 1:two_third])])
sns.countplot(x=predicted, ax=ax[1]);
predicted = kmeans.predict(scaled_X[df[col].isin(_counts.index[two_third + 1:_cumsum[-1]])])
sns.countplot(x=predicted, ax=ax[2]);
fig.show()
```

Looking ath the cluster distribution against low, medium and high counts of patients in PCNs we crealy see some different distributions. This indicates that we can have latent space or reduce the dimensionality while we still can explain our data.

If I had more time it would be interesting to try similarly against other variables. And other clustering techniques, it can be seen here that sizes of clusters are also not of equal size.

### Dimensionality Reduction (Principal Component Analysis)

As seen above dimensionality reduction may be able to explain our data with reducing complexity.

```python
from sklearn.decomposition import PCA
```

```python
'''
Fitting PCA model and transforming our data.
'''
n_components=X.shape[1]
pca = PCA(n_components=n_components)
pca.fit(scaled_X)
X_pca = pca.transform(scaled_X)
```

```python
'''
Printing and plotting variance explained against number of principal components.
'''
var_explained = pd.DataFrame({'Variance Explained':pca.explained_variance_ratio_.cumsum()})
var_explained.index = np.array(var_explained.index) + 1
var_explained.index.names = ['N components']
print(var_explained)
pca_df = pd.DataFrame({'Explained Variance':pca.explained_variance_ratio_.cumsum(), 'N components': range(0, pca.n_components)})
sns.lineplot(x='N components', y='Explained Variance', data=pca_df);
```

We that the first 9 principal components are able to explain 94% of the variance in our dataset, allowing for a reduction of 11 features.
Benefits from dimensionality reduction include:
- Reducing affects of curse of dimensionality
- Reducing size and processing load
- Removal of some multicollinearity, improving interpretation of features for ML purposes.

```python
'''
Fitting PCA model with 2 principal components.
'''
n_components=2
pca = PCA(n_components=n_components)
pca.fit(scaled_X)
X_pca = pca.transform(scaled_X)
var_explained = pd.DataFrame({'Variance Explained':pca.explained_variance_ratio_.cumsum()})
var_explained.index = np.array(var_explained.index) + 1
var_explained.index.names = ['N components']
print(var_explained)
```

```python
col = '2018-19Register'
print(col)
prev_large = scaled_X[col] > scaled_X[col].mean() # create binary indicator for when '2018-19Register' larger than mean
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=prev_large).set(xlabel='Component 0', ylabel='Component 1');
```

```python
col = '2018-19Achievement (%)'
print(col)
prev_large = scaled_X[col] > scaled_X[col].mean() # create binary indicator for when '2018-19Register' larger than mean
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=prev_large).set(xlabel='Component 0', ylabel='Component 1');
```

Fitting the PCA with a limit of only 2 components we see that only %48 of the variance is explained. But with the lower dimension representation we are able to easily visualise, as above with the 2D scatterplot.

In this plot I colour the points for when '2018-19Register' is larger than mean to Orange, equal and less to Blue. In our case we see a clear pattern of positive relationship will '2018-19Register' and 'Component 0' showing it as an important feature to explain to data. Plotting similarly with '2018-19Achievement (%)' we see less of a pattern.
