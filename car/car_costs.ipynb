```python
import seaborn as sns
import pandas as pd
import datetime
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
def gen_df(start_date, end_date, same_3, used_3, used_price, finance):
    dates = [start_date]
    sames = [same_3]
    useds = [used_3]
    finances = [finance]
    while dates[-1] <= end_date:
        dates.append(dates[-1] + datetime.timedelta(days=7))
        sames.append(sames[-1] + same_3)
        useds.append(useds[-1] + used_3)
        finances.append(finances[-1] + finance)

    df = pd.DataFrame({'date': dates, 'keep_cost': sames, 'used_cost': useds, 'finance': finances})
    df['used_cost'] = df['used_cost'] + used_price
    return df
```

# For 3 Day a week commute

```python
same_3 = 118
used_3 = 62.2
used_price = 15000
finance = 159.375

start_date = datetime.datetime(2022, 9, 1)
end_date = datetime.datetime(2027, 9, 1)
```

```python
df = gen_df(start_date, end_date, same_3, used_3, used_price, finance)

plot_df = pd.melt(df, id_vars=['date'])
sns.lineplot(data=plot_df, x='date', y='value', hue='variable', palette='colorblind')
```
